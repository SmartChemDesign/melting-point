 **Calculation Morgan fingerprints and usage xgb**

```python
import time
t0start = time.time() 

from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('df.csv')
df
```

```python
%%time
fingerprints = []
    
from rdkit.Chem import AllChem
from rdkit.Chem import rdFingerprintGenerator

# Create a Morgan fingerprint generator
morgan_generator = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)
for smi in df["SMILES"]:
    mol = Chem.MolFromSmiles(smi)
    fp = morgan_generator.GetFingerprint(mol) 
    arr = np.array(fp)
    fingerprints.append(arr)
```

```python
len(fingerprints[0])
```

```python
X_morgan = pd.DataFrame(np.array(fingerprints))
X_morgan
```

```python
%%time
fingerprints3 = []

morgan_generator = rdFingerprintGenerator.GetMorganGenerator(radius=3, fpSize=2048)
for smi in df["SMILES"]:
    mol = Chem.MolFromSmiles(smi)
    fp = morgan_generator.GetFingerprint(mol) 
    arr = np.array(fp)
    fingerprints3.append(arr)
```

```python
len(fingerprints3[0])
```

```python
X_morgan3 = pd.DataFrame(np.array(fingerprints3))
X_morgan3
```

```python
emb_morgan_df = pd.concat([X_morgan, X_morgan3], ignore_index=True, axis=1)
emb_morgan_df
```

```python
mps = [mp for mp in df["MP"]]
len(mps)
```

```python
emb_morgan_df["MP"] = mps
emb_morgan_df
```

```python
emb_morgan_df.to_csv('emb_morgan_df.csv', index=False)
```

```python
emb_morgan_df = pd.read_csv('emb_morgan_df.csv')
emb_morgan_df
```

```python
from sklearn.model_selection import train_test_split
X_mor_train, X_mor_test, Y_train, Y_test = train_test_split(emb_morgan_df.iloc[:, :-1], emb_morgan_df.iloc[:, -1], test_size=0.2, random_state=42)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
%%time
model = xgb.XGBRegressor(n_estimators=1000, n_jobs=-1, learning_rate=0.01, max_depth=7)

model.fit(X_mor_train, Y_train)

Y_pred = model.predict(X_mor_test)

mse = mean_squared_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = mse ** 0.5
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"MSE: {mse}, RMSE: {rmse}, MAE: {mae}, R²: {r2}")
```

```python
print(f"MSE: {mse:.2f}, RMSE: {rmse:.2f}, MAE: {mae:.2f}, R²: {r2:.2f}")
```

```python
%%time

import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error

X = emb_morgan_df.iloc[:, :-1]
y = emb_morgan_df.iloc[:, -1]

n_folds = 5
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

maes = []
rmsses = []

for train_idx, val_idx in kf.split(X):
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_val, y_val = X.iloc[val_idx], y.iloc[val_idx]
    
    model = xgb.XGBRegressor(
        n_estimators=1000,
        n_jobs=-1,
        learning_rate=0.05,
        max_depth=7
    )
    model.fit(X_train, y_train)

    y_pred = model.predict(X_val)
    
    mse = mean_squared_error(y_val, y_pred)
    rmse = np.sqrt(mse)
    
    mae = mean_absolute_error(y_val, y_pred)

    rmsses.append(rmse)
    maes.append(mae)

mean_mae = np.mean(maes)
std_mae = np.std(maes)
mean_rmse = np.mean(rmsses)
std_rmse = np.std(rmsses)

print(f'MAE: {mean_mae:.4f} ± {std_mae:.4f}')
print(f'RMSE: {mean_rmse:.4f} ± {std_rmse:.4f}')
```
